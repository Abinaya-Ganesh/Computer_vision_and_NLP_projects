**EVALUATION METRICS**

```python
#Evaluation metrics Dataframe
import pandas as pd
import os

path = r"C:\Users\abina\Desktop\abi\final project files\Final project\csv files\evaluation_metrics"
files = os.listdir(path)

df_metrics = pd.concat([pd.read_csv(path+"\\"+file) for file in files], ignore_index=True)
```

```python
df_metrics = df_metrics[['Dataset','Model name','Accuracy','Precision','Recall','F1-score']]
df_metrics = df_metrics.round({'Precision':2,'Recall':2,'F1-score':2})

iterables = [["CIFAR_10", "FASHION MNIST", "MNIST"], ["LeNet5", "AlexNet", "VGG", "ResNet18", "SENet18", "GoogleNet"]]
index = pd.MultiIndex.from_product(iterables, names=["Dataset", "Model name"])
df_metrics1 = df_metrics.set_index(keys=index)

df_metrics1.drop(columns = ['Dataset','Model name'], inplace=True)
df_metrics1
```

**MNIST dataset evaluation metrics**

```python
print("MNIST dataset evaluation metrics")
df1 = df_metrics[['Model name','Accuracy','Precision','Recall','F1-score']][df_metrics['Dataset'] == 'MNIST']
df1.set_index(keys="Model name",inplace=True)

df1.style.format('{:.2f}').highlight_max(color = 'lightgreen').highlight_min(color = 'coral')
```

**Fashion MNIST dataset evaluation metrics**

```python
print("Fashion MNIST dataset evaluation metrics")
df1 = df_metrics[['Model name','Accuracy','Precision','Recall','F1-score']][df_metrics['Dataset'] == 'FASHION MNIST']
df1.set_index(keys="Model name",inplace=True)
df1.style.format('{:.2f}').highlight_max(color = 'lightgreen').highlight_min(color = 'coral')
```

**CIFAR-10 dataset evaluation metrics**

```python
print("CIFAR-10 dataset evaluation metrics")
df1 = df_metrics[['Model name','Accuracy','Precision','Recall','F1-score']][df_metrics['Dataset'] == 'CIFAR-10']
df1.set_index(keys="Model name",inplace=True)
df1.style.format('{:.2f}').highlight_max(color = 'lightgreen').highlight_min(color = 'coral')
```
