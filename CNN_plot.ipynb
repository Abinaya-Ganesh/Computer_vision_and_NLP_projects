**EVALUATION METRICS**

**MNIST dataset evaluation metrics**

```python
path = r"C:\Users\abina\Desktop\abi\loss and metrics\Metrics\MNIST"
files = os.listdir(path)
m_metrics = []

for file in files:
    m_metrics.append(pickle.load(open(path + "\\" + file, "rb")))
```

```python
df_m_metrics = pd.DataFrame(m_metrics)

df_m_metrics.loc[3,'Model name'] = 'VGG_modified'

df_m_metrics = df_m_metrics[['Dataset','Model name','Accuracy','Precision','Recall','F1-score']]
df_m_metrics.drop(columns = ['Dataset'], inplace=True)
df_m_metrics.set_index(keys="Model name",inplace=True)

print("MNIST dataset evaluation metrics")
df_m_metrics.style.format('{:.2f}').highlight_max(color = 'lightgreen').highlight_min(color = 'coral')
```
